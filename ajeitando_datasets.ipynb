```python
import pandas as pd
dados = pd.read_csv('/content/ctd_final.csv')

# cortando a primeira e última linha do dado
dados = dados[1:1499]

# colocando os nomes das colunas em minúsculo
dados = dados[dados.columns[0:11]].set_axis([nomes.lower() for nomes in dados.columns[0:11]], axis=1, inplace=False) 

# transformando para numérico
dados[dados.columns[1:11]] = dados[dados.columns[1:11]].apply(pd.to_numeric, errors='coerce', axis=1) 

# arredondando os valores
dados['ctdsal'] = [round(x,2) for x in dados.ctdsal]
dados['ctdtmp'] = [round(x,2) for x in dados.ctdtmp]

dados
```

```python
estacao = pd.read_csv('/content/estacao_rj.csv')

# colocando os nomes para minuscula
estacao = estacao[estacao.columns[:]].set_axis([nomes.lower() for nomes in estacao.columns[:]], axis=1, inplace=False)

# setando um index datetime
estacao.index = pd.to_datetime((estacao['ano'].astype(int).astype(str)+'/'+estacao['mes'].astype(int).astype(str)+'/'+estacao['dia'].astype(int).astype(str)))
estacao.drop(['ano','mes','dia'], axis=1, inplace=True)

estacao
```

```python
boia = pd.read_csv('/content/SIMCOSTA_RJ-4_OCEAN_2017-08-28_2021-11-12.csv', skiprows=21)

# nomes das colunas em minúsculo
boia = boia[boia.columns[:]].set_axis([nomes.lower() for nomes in boia.columns[:]], axis=1, inplace=False)

# colocando índice datetime
boia.index = pd.to_datetime(boia['year'].astype(str)+'-'+boia['month'].astype(str)+'-'+boia['day'].astype(str)+'-'+boia['hour'].astype(str)+':'+boia['minute'].astype(str))
boia.drop([nomes for nomes in boia.columns[0:7]], axis=1, inplace=True)

# removendo colunas que tenham mais de 50% de valores faltantes
for nome in boia.columns[:]:
  if boia[nome].isnull().sum()/len(boia[nome]) > 0.5: 
    boia.drop(nome, axis=1,inplace=True)

boia
```

```python
import pandas as pd
nino = pd.read_csv('/content/elnino.csv')

# colocando o titulo das colunas para minusculo
nino = nino[nino.columns[:]].set_axis([nome.lower() for nome in nino.columns[:]], axis=1, inplace=False)

# tirando os espaços dos nomes das colunas
nino = nino[nino.columns[:]].set_axis([nome.strip() for nome in nino.columns[:]], axis=1, inplace=False)

# colocando o índice em datetime
nino.index = pd.to_datetime(nino['year'].astype(str)+'-'+nino['month'].astype(str)+'-'+nino['day'].astype(str))

# removendo as colunas que não vou usar
nino.drop([nomes for nomes in nino.columns[0:5]], axis=1, inplace=True)

# transformando as colunas em numérico
nino = nino[nino.columns[:]].apply(pd.to_numeric, errors='coerce', axis=1)

# removendo colunas que tenham mais de 30% de valores faltantes
for nome in nino.columns[:]:
  if nino[nome].isnull().sum()/len(nino[nome]) > 0.3: 
    nino.drop(nome, axis=1,inplace=True)

nino
```
